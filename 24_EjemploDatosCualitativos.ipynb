## Juntar color de ojos y pelo sin distinguir por sexo

```r
HairEyeColor
```

```r
summary(HairEyeColor)
```

```r
str(HairEyeColor)
```

```r
male <- HairEyeColor[,,"Male"]  # Extraer la matriz de datos de los hombres
male
```

```r
female <- HairEyeColor[,,"Female"]
female
```

```r
data <- as.table(male+female) # Sumar las matrices de hombres y mujeres
data
```

```r
dimnames(data) = list(
    Pelo = c("Negro", "Marron", "Pelirrojo", "Rubio"),
    Ojos = c("Marrones", "Azules", "Pardos", "Verdes")
)
data
```

## Diagrama de mosaico

```r
plot(data, col = c("lightblue"), main = "Diagrama de Mosaico")
```

## calcular el numero total de individues 

```r
sum(data) #
```

```r
colSums(data) 
```

```r
rowSums(data)
```

```r
round(prop.table(colSums(data)),3)
```

```r
round(prop.table(rowSums(data)),3)
```

## Diagrama de barras:

```r
barplot(prop.table(colSums(data)), ylim = c( 0, 0.4), 
            main = "Frecuencias relativas del color de ojos",
            col = c("burlywood4", "lightblue", "orange3", "lightgreen")
        )
```

## Frecuencias relativas globales y marginales:

```r
round(prop.table(data),3) # Frecuencias relativas globales
```

```r
round(prop.table(data, margin = 1),3) # Frecuncias relativas marginales por color de pelo, para cada color de ojos
```

```r
round(prop.table(data, margin = 2),3) # Frecuencias relativas marginales por color de ojos, para cada color de pelo
```

```r
barplot(prop.table(data, margin = 1), beside = TRUE,
        legend.text = TRUE,
        ylim = c(0, 0.8),
        col = c("black", "burlywood4", "red", "gold"),
        main = "Frecuncias relativas del color de pelo \n para cada color de ojo."
)
```

```r
barplot(t(prop.table(data, margin = 2)), beside = TRUE,
        legend.text = TRUE,
        ylim = c(0, 0.6),
        col = c("burlywood4", "lightblue", "orange3", "lightgreen"),
        main = "Frecuncias relativas del color de ojos \n para cada color de pelo."
)
```
