## Diagrama de barras

```r
library(gmodels) # Cargamos la librería

# Tabla de Frecuencias Relativas:
Sexo = factor(c("H", "M", "M", "M", "H", "H", "M", "M"))
Respuestas = factor(c("Si", "No", "Si", "Si", "Si", "No", "No", "Si"))
CrossTable(Sexo, Respuestas, prop.chisq = FALSE)
```

```r
barplot(table(Sexo), col = c("lightblue", "pink"), main = "Diagrama de barras de
las frecuencias absolutas \n de la variable \"Sexo\"")
```

```r
barplot(prop.table(table(Respuestas)), # Frecuencias relativas
    main = "Diagrama de barras de frecuencias relativas \n de la variable \"Respuestas\""
)
```

```r
par(mfrow = c(1, 2)) # Dividimos la ventana gráfica en dos
barplot(table(Respuestas), col = c("green"))
barplot(table(Respuestas), col = c("red", "blue"))
```

```r
barplot(table(Sexo), horiz = TRUE)
```

```r
barplot(table(Sexo,Respuestas),legend.text = TRUE) # Diagrama de barras apiladas de las frecuencias absolutas de las variables "Sexo" y "Respuestas" con legendas
```

```r
barplot(table(Sexo, Respuestas), beside = TRUE, legend.text = TRUE)
```

```r
barplot(table(Respuestas, Sexo), beside = TRUE, names=c("Men", "Women"),
    col = c("red", "yellow"), legend.text = c("Negativo", "Positivo")
)
```
