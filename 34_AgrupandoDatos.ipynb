### Funciones para las tablas de frecuencias:

```r
#Primera funcin
TablaFrecs = function(x,k,A,p){ 
  L = min(x)-p/2+A*(0:k)
  x_cut = cut(x, breaks = L, right=FALSE)
  intervals = levels(x_cut) 
  mc = (L[1]+L[2])/2+A*(0:(k-1))
  Fr.abs = as.vector(table(x_cut)) 
  Fr.rel = round(Fr.abs/length(x),4) 
  Fr.cum.abs = cumsum(Fr.abs) 
  Fr.cum.rel = cumsum(Fr.rel)
  tabla = data.frame(intervals, mc, Fr.abs, Fr.cum.abs, Fr.rel, Fr.cum.rel)
  tabla
  }
```

```r
TablaFrecs.L = function(x,L,V){
  x_cut = cut(x, breaks=L, right=FALSE, include.lowest=V)
  intervals = levels(x_cut)
  mc = (L[1:(length(L)-1)]+L[2:length(L)])/2
  Fr.abs = as.vector(table(x_cut)) 
  Fr.rel = round(Fr.abs/length(x),4)
  Fr.cum.abs = cumsum(Fr.abs)
  Fr.cum.rel = cumsum(Fr.rel)
  tabla = data.frame(intervals, mc, Fr.abs, Fr.cum.abs, Fr.rel, Fr.cum.rel)
  tabla
  }
```

```r
notas = c(7 ,10 , 2, 2, 6, 2, 5, 4, 9, 2, 7, 5, 1, 7, 0, 3, 10, 2, 10, 4, 1, 4, 5, 4, 0,
          5, 10,  4, 3, 0, 7, 5, 10, 3, 4, 8, 1, 9, 3, 7, 9, 1, 9, 10, 5, 10, 10, 9, 5, 0,
          3, 1, 3, 2, 0, 6, 6, 4, 7, 4, 7, 3, 9, 0, 7, 0, 3, 0, 3, 3, 1, 4, 10, 9, 1,
          4, 0, 6, 10, 0, 10, 1, 0, 2, 6, 4, 8, 2, 3, 7, 7, 3, 3, 8, 2, 6, 6, 2, 8, 9
        )
notas
```

```r
L = c(0, 5, 7, 9, 10) #Definimos el vector de extremos para los intervalos

#creamos los intervalos:
notas1 = cut(notas, #data
            breaks = L, # cortes para los intervalos
            right = FALSE, # cerrado por la derecha, abierto por la izquierda
            include.lowest = TRUE) # incluye el ultimo dato
notas1
```

```r
# Definimos las marcas de clase
MC = (L[1:length(L)-1]+L[2:length(L)])/2
notas2 = cut(notas, breaks = L, labels = MC, right = FALSE, include.lowest = TRUE)
notas2
```

```r
#Definimos notas3 como el resultado de la codificacin en intervalos utilizando como 
#etiquetas la posicin ordenada del intervalo (1, 2, 3 o 4)
notas3 = cut(notas, breaks = L, labels = FALSE, right = FALSE, include.lowest = TRUE)
notas3
```

```r
#Definimos notas4 como el resultado de la codificacin en intervalos utilizando como 
#etiquetas Susp, Aprob, Not y Exc
notas4 = cut(notas, breaks = L, 
            labels = c("Susp", "Aprob", "Not", "Exc"), 
            right = FALSE, 
            include.lowest = TRUE)
notas4
```

```r
# de la data notas4:
table(notas4) #Frecuencia absoluta
```

```r
cumsum(table(notas4)) #Frecuencia absoluta acumulada
```

```r
prop.table(table(notas4)) #Frecuenca relativa
```

```r
cumsum(prop.table(table(notas4))) #Frecuenca relativa acumulada
```

```r
notasHist = hist(notas, breaks = L, right = FALSE, include.lowest = TRUE, plot = FALSE)
notasHist
```

```r
FAbs = notasHist$count
FAbs
```

```r
FRel = prop.table(FAbs)
FRel
```

```r
FAbsCum = cumsum(FAbs)
FAbsCum
```

```r
FRelCum = cumsum(FRel)
FRelCum
```

```r
intervalos = c("[0,5)","[5,7)","[7,9)","[9,10]")
calificacion = c("Suspenso", "Aprobado", "Notable", "Excelente")
marcas = notasHist$mids
tabla.Fr = data.frame(intervalos,calificacion,marcas,FAbs,FAbsCum,FRel,FRelCum)
tabla.Fr
```
