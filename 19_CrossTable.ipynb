```r
library(gmodels) # Cargamos la librera
```

```r
# Tabla de Frecuencias Relativas:
sex = factor(c("H", "M", "M", "M", "H", "H", "M", "M"))
ans = factor(c("S", "N", "S", "S", "S", "N", "N", "S"))
CrossTable(sex, ans, prop.chisq = FALSE)
```

```r
# Sumas por filas y columnas

tt <- table(sex, ans)
tt
```

```r
# marginales globales
prop.table(tt) # Frecuencias relativas globales
```

```r
# por filas (sexo)
prop.table(tt, margin = 1) # por filas (frecuencias relativas por sexo)
```

```r
# por columnas (ans)
prop.table(tt, margin = 2) # por columnas (frecuencias relativas por ans)
```

```r
rowSums(tt) # Suma por filas
```

```r
# otro metodo para sumar filas
apply(tt, FUN = sum, MARGIN = 1) # Suma por filas
```

```r
colSums(tt) # Suma por columnas
```

```r
# otro metodo para sumar columnas
apply(tt, FUN = sum, MARGIN = 2) # Suma por columnas
```

```r
colSums(prop.table(tt)) # Suma por columnas de las frecuencias relativas
```

```r
rowSums(prop.table(tt)) # Suma por filas de las frecuencias relativas
```
