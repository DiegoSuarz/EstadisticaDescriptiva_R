```r
# Funcion de probabilidad (binomial)
# X = B(n = 30, p = 0.6)
n = 30
p = 0.6
dbinom(0:30, size = n, prob = p)
```

```r
plot(0:30,dbinom(0:30, size = n, prob = p))
```

```r
library(Rlab)
plot(0:n, dbinom(0:n, size = n, prob = p))
```

```r
plot(0:n, pbinom(0:n, size = n, prob = p))
```
