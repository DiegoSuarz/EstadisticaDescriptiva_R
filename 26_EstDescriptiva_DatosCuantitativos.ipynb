## Ejemplo 1:

```r
edad = c(15,18,25,40,30,29,56,40,13,27,42,23,11,26,25,32,30,40,33,29)
edad
```

```r
# Frecuencias absolutas
table(edad)
```

```r
# frecuencia absoluta acumulada
cumsum(table(edad))
```

```r
# Frecuencia relativa
round(prop.table(table(edad)),3)
```

```r
# Frecuencia relativa acumulada
round(cumsum(prop.table(table(edad))),3)
```

## Ejemplo 2

```r
set.seed(162017) # Semilla para reproducibilidad
dados = sample(1:6, 25, replace = TRUE) # Lanzamiento de 25 dados, replace = TRUE permite repetir los lanzamientos
dados
```

```r
set.seed(NULL)
```

```r
# Frecuencias absolutas
table(dados)
```

```r
# Frecuencia absoluta acumulada
cumsum(table(dados))
```

```r
# frecuencia relativa
round(prop.table(table(dados)),2)
```

```r
# Frecuencia relativa acumulada
round(cumsum(prop.table(table(dados))),2)
```

```r
dados.df = data.frame(Puntuacion = 1:6,
                        Fr.abs = as.vector(table(dados)),
                        Fr.abs.acum = as.vector(cumsum(table(dados))),
                        Fr.rel = as.vector(round(prop.table(table(dados)),2)),
                        Fr.rel.acum = as.vector(round(cumsum(prop.table(table(dados))),2))
                        )   
dados.df
```
