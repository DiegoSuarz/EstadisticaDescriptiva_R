## Tabla de frecuencias unidimensionales:

```r
# Tablas de contingencia

datos = factor(c("H", "M", "M", "M", "H", "H", "M", "M"))
table(datos) 
```

```r
table(datos)["M"]
```

```r
table(datos)[2]
```

```r
sum(table(datos))
```

```r
# Tabla de Frecuencias Relativas:
datos = factor(c("H", "M", "M", "M", "H", "H", "M", "M"))
prop.table(table(datos))
```

```r
# Tabla de Frecuencias Relativas porcentuales:  
100*prop.table(table(datos))
```

```r
table(datos)/length(datos)
```

```r
names(table(datos))
```

```r
names(which(table(datos) == 3))
```

```r
names(which(table(datos) == max(table(datos))))
```

```r
# Calcular la moda:
moda <- function(d){
    names(which(table(d) == max(table(d))))
}

moda(datos)
```
