# Medidas de posicion

## Ejercicios:

```r
set.seed(260798)
dado = sample(1:4, 50, replace = TRUE)
set.seed(NULL)
length(dado)
```

```r
dado = sort(dado)  # Ordenamos de menor a mayor
dado
```

```r
df.dado = data.frame(Puntuacion = 1:4,
                    Fr.abs = as.vector(table(dado)),
                    Fr.abs.acu = as.vector(cumsum(table(dado))),
                    Fr.rel = as.vector(round(prop.table(table(dado)),2)),
                    Fr.rel.acum = as.vector(round(cumsum(prop.table(table(dado))),2))
                    )
df.dado
```

```r
# el Q30 seria el primer elemento de dicha lista tal que fuera mayor o igual que el 30% 
# de los datos, si calculamos el 30% de 50 (total de datos), obtenemos que se 15, el 
# numero que buscamos esta en la posicion 15 de esa lista.
dado[15]
```

```r
quantile(dado,0.3, type = 7)
```
