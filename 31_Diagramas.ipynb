```r
data = read.table("base de datos/datacrab.txt", header = TRUE)
boxplot(weight~color, data = data)
```

```r
medias = aggregate(weight~color, data = data, FUN = mean)
medias

boxplot(weight~color, data = data, notch= TRUE)
points(medias, col= "blue", pch = 21) #dibujar las medias
```

```r
str(boxplot(weight~color, data = data))
```

```r
boxplot(weight~color, data = data, notch= TRUE)$stats #min, 1r quartil, mediana, 3er quartil, max
```

```r
boxplot(weight~color, data = data, notch= TRUE)$out #outliers
```

```r
boxplot(weight~color, data = data, notch= TRUE)$group
```
