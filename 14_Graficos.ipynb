```r
## Parametros:

n = 1:20
fib = (1/sqrt(5))*(((1+sqrt(5))/2)^n - ((1-sqrt(5))/2)^n)
fib 

plot(fib, #datos a graficar
    xlab = "n", #nombre del eje x
    ylab = expression(F[n]), #nombre del eje y
    main = "Sucesion de Fibonacci", #titulo del grafico
    pch = 21, #tipo de punto
    cex =3, #tamaño del punto
    col = "powderblue", bg = "orchid", #color del borde y del relleno
    log = "xy") #escala logaritmica en ambos ejes
```

```r
par(mfrow = c(1,2)) #divide la ventana en 1x2
plot(fib, #datos a graficar
    xlab = "n", #nombre del eje x
    ylab = expression(F[n]), #nombre del eje y
    main = "Sucesion de Fibonacci", #titulo del grafico
    pch = 21, #tipo de punto
    cex =2, #tamaño del punto
    col = "powderblue", bg = "orchid", #color del borde y del relleno
    log = "xy") #escala logaritmica en ambos ejes

plot(fib, #datos a graficar
    xlab = "n", #nombre del eje x
    ylab = expression(F[n]), #nombre del eje y
    main = "Sucesion de Fibonacci 2", #titulo del grafico
    pch = 21, #tipo de punto
    cex =2, #tamaño del punto
    col = "powderblue", bg = "orchid", #color del borde y del relleno
    log = "y") #escala logaritmica en el eje y
par(mfrow = c(1,1)) #vuelve a la ventana normal
```

```r
f <- function(x){x^2}
plot(f, xlim = c(-3,3)) 
points(-3:3, f(-3:3), pch = 19)
lines(-3:3, #coordenadas x
    f(-3:3)*1.5, #coordenadas y
    lwd = 2, #grosor de la linea
    lty = "dotted", #tipo de linea
    col = "red") #color de la linea
```

```r
f <- function(x){x^2}
plot(f, xlim = c(-3,3), ylim = c(-10,10)) 
points(-3:3, f(-3:3), pch = 19)
curve(x^3, #funcion a graficar
    lty = "dashed", #tipo de linea
    col = "blue", #color de la linea
    add = TRUE) #agrega la curva al grafico
```

```r
x = seq(0,2*pi, 0.1) #genera un vector de 0 a 2pi con paso 0.1
plot(x, sin(x), type = "l", col = "blue", lwd = 3, xlab = "", ylab = "")
lines(x, cos(x), col = "green", lwd = 3)
lines(x, tan(x), col = "purple", lwd =3)

legend("bottomleft", #posicion de la leyenda
    col = c("blue", "green", "purple"), #colores de las lineas
    legend = c("Seno", "Coseno", "Tangente"), #nombres de las lineas
    lwd = 3, #grosor de las lineas
    bty = "l") #tipo de caja

```

```r
f <- function(x){x^2}
plot(f, xlim = c(-3,3), ylim = c(-10,10)) 
points(-3:3, f(-3:3), pch = 19)
curve(x^3, #funcion a graficar
    lty = "dashed", #tipo de linea
    col = "blue", #color de la linea
    add = TRUE) #agrega la curva al grafico

curve(x^4, lty = "dashed", col = "orangered", add = TRUE)

legend("bottomright", #posicion de la leyenda
    legend = c(expression(x^2), expression(x^3), expression(x^4)), #nombres de las lineas # nolint
    lwd = 2, 
    col = c("red","blue","orangered"), #colores de las lineas
    lty = c("dotted", "dashed", "dashed") #tipo de linea
    )
```

```r
x = c(5*(1:10))
plot(x, c(exp(-x)+(-1)^x*x/2*sin(x)^2), xlab = "", ylab = "", main = "Grafico con varios elementos") # nolint

# Agregamos una linea
segments(10, #endregion
        0, #coordenada y inicial # nolint: indentation_linter.
        40, #coordenada x final
        0, #coordenada y final
        col = "red", #color de la linea
        lwd = 4 #grosor de la linea
        )

# Agregamos una flecha
arrows(10, #coordenada x inicial
        0, #coordenada y inicial
        40, #coordenada x final
        -10, #coordenada y final
        col = "blue", #color de la flecha
        length = 0.5, #longitud de la flecha
        angle = 5,  #angulo de la flecha
        code = 3 #tipo de flecha
) 

# Agregamos una estrella
symbols(40, #coordenada x 
        0, #coordenada y 
        stars = cbind(1, .5, 1, .5, 1, .5, 1, .5, 1, .5), #coordenadas de inicio de las lineas # nolint: line_length_linter.
        add = TRUE, #agrega al grafico
        lwd = 3, #grosor de las lineas
        inches = 0.5 #tamaño de las lineas
        )
# Agregamos una estrella
symbols(40, #coordenada x 
        0, #coordenada y 
        stars = cbind(1, .5, 1, .5, 1, .5, 1, .5, 1, .5), #coordenadas de inicio de las lineas # nolint: line_length_linter.
        add = TRUE, #agrega al grafico
        lwd = 3, #grosor de las lineas
        )

# agregamos un poligono
polygon(c(20,30,40), #coordenadas x
    c(10,-10,10), #coordenadas y
    col = "gold", #color del poligono
    density = 3, #densidad de lineas
    angle = 90, #angulo de las lineas
    lty = 4, #tipo de linea
    lwd = 5 #grosor de la linea 
    )
```
