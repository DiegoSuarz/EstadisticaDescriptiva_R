```r
x[c(1, 5, 6)] # elementos 1, 5 e 6
```

```r
x[x > 30] # elementos mayores que 30
```

```r
x[x>20 & x<40] # elementos entre 20 e 40
```

```r
x[x != 3 & x != 17] # elementos diferentes de 3 e 17
```

```r
x[x < 10 | x > 40] # elementos menores que 10 o mayores que 40
```

```r
x[x > 10] # elementos mayores que 10
```

```r
x[x >= 10]  # elementos mayores o iguales a 10
```

```r
x[!x < 10] # elementos mayores o iguales a 10
```

```r
x[x %% 2 == 0] # elementos pares            
```

```r
x[x %% 2 == 1] # elementos impares
```
